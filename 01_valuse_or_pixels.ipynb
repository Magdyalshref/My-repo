```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr=np.arange(0,256,1)
print(arr)
```

```python
arr=arr.reshape((16,16))
print(arr)
```

```python
arr_8bit=arr.astype(np.uint8)
```

```python
plt.imshow(arr_8bit,cmap='gray')
plt.axis('off')
plt.show()
```

```python
arr1=np.random.randint(0,255,(16,16),np.uint8)
plt.imshow(arr1,cmap='gray')
plt.axis('off')
```
